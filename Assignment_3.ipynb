### 1. Defanging an IP Address

```python
def defanging(ip_address):
    return ip_address.replace('.','[.]')
    
ip_address = input("Enter your ip address:")
defanging(ip_address)
```

```python
def defangingIpAddress(address):
    out = []
    for ch in address:
        if ch == '.':
            out.append('[.]')
        else:
            out.append(ch)
    return ''.join(out)

address = input("Enter your IP Address:")
defangingIpAddress(address)
```

### 2. Find Numbers with Even Number of Digits

```python
def findNumbers(nums):
    count = 0
    for n in nums:
        if len(str(n)) % 2 == 0:
            count += 1

    return count

nums = [12, 345, 2, 6, 7896]
findNumbers(nums)
```

### 3. Number of Good Pairs

```python
def goodPairs(nums):
    count = 0

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                count += 1

    return count

nums = [1,2,3,1,1,3]
goodPairs(nums)
```

### 4. How Many Numbers Are Smaller Than the Current Number

```python
def smallerNumsCount(nums):
    res = []

    for i in range(len(nums)):
        count = 0
        for j in range(len(nums)):
            if nums[i] > nums[j]:
                count += 1
        res.append(count)

    return res

nums = [8,1,2,2,3]
smallerNumsCount(nums)
```

### 5. Subtract the Product and Sum of Digits of an Integer

```python
def subProdAndSum(num):
    digits = [int(d) for d in str(num)]

    product = 1
    for d in digits:
        product *= d

    total = sum(digits)
    sub = product - total

    return sub

nums = int(input("Enter your number:"))
subProdAndSum(nums)
```

### 6. XOR Operation in an Array

```python
def xorOperation(n, start):
    nums = []

    for i in range(n):
        nums.append(start + 2 * i)

    res = 0
    for num in nums:
        res ^= num

    return res

n = int(input("Enter your number:"))
start = int(input("Enter your start number:"))
xorOperation(n, start)
```
